```python
%%writefile trader.py

############################################
# state: TradingState parameter
############################################

import json
from typing import Any
from datamodel import Listing, Observation, Order, OrderDepth, ProsperityEncoder, Symbol, Trade, TradingState

class Logger:
    def __init__(self) -> None:
        self.logs = ""
        self.max_log_length = 3750

    def print(self, *objects: Any, sep: str = " ", end: str = "\n") -> None:
        self.logs += sep.join(map(str, objects)) + end

    def flush(self, state: TradingState, orders: dict[Symbol, list[Order]], conversions: int, trader_data: str) -> None:
        base_length = len(
            self.to_json([
                self.compress_state(state, ""),
                self.compress_orders(orders),
                conversions,
                "",
                "",
            ])
        )

        max_item_length = (self.max_log_length - base_length) // 3

        print(
            self.to_json([
                self.compress_state(state, self.truncate(state.traderData, max_item_length)),
                self.compress_orders(orders),
                conversions,
                self.truncate(trader_data, max_item_length),
                self.truncate(self.logs, max_item_length),
            ])
        )

        self.logs = ""

    def compress_state(self, state: TradingState, trader_data: str) -> list[Any]:
        return [
            state.timestamp,
            trader_data,
            self.compress_listings(state.listings),
            self.compress_order_depths(state.order_depths),
            self.compress_trades(state.own_trades),
            self.compress_trades(state.market_trades),
            state.position,
            self.compress_observations(state.observations),
        ]

    def compress_listings(self, listings: dict[Symbol, Listing]) -> list[list[Any]]:
        return [[listing.symbol, listing.product, listing.denomination] for listing in listings.values()]

    def compress_order_depths(self, order_depths: dict[Symbol, OrderDepth]) -> dict[Symbol, list[Any]]:
        return {
            symbol: [order_depth.buy_orders, order_depth.sell_orders]
            for symbol, order_depth in order_depths.items()
        }

    def compress_trades(self, trades: dict[Symbol, list[Trade]]) -> list[list[Any]]:
        compressed = []
        for arr in trades.values():
            for trade in arr:
                compressed.append([
                    trade.symbol,
                    trade.price,
                    trade.quantity,
                    trade.buyer,
                    trade.seller,
                    trade.timestamp,
                ])
        return compressed

    def compress_observations(self, observations: Observation) -> list[Any]:
        conversion_observations = {}
        for product, observation in observations.conversionObservations.items():
            conversion_observations[product] = [
                observation.bidPrice,
                observation.askPrice,
                observation.transportFees,
                observation.exportTariff,
                observation.importTariff,
                observation.sugarPrice,
                observation.sunlightIndex,
            ]
        return [observations.plainValueObservations, conversion_observations]

    def compress_orders(self, orders: dict[Symbol, list[Order]]) -> list[list[Any]]:
        compressed = []
        for arr in orders.values():
            for order in arr:
                compressed.append([order.symbol, order.price, order.quantity])
        return compressed

    def to_json(self, value: Any) -> str:
        return json.dumps(value, cls=ProsperityEncoder, separators=(",", ":"))

    def truncate(self, value: str, max_length: int) -> str:
        return value[:max_length]

logger = Logger()

MAX_VOLUME = 80

class PepperTrader:
    def long_only(order_depth, state, best_bid, orders, curr_pepper_position):
        if curr_pepper_position < MAX_VOLUME:
            remaining = MAX_VOLUME - curr_pepper_position

            for price, volume in sorted(order_depth.sell_orders.items()):
                if remaining <= 0:
                    break

                available_qty = -volume
                buy_qty = min(available_qty, remaining)

                orders.append(Order("INTARIAN_PEPPER_ROOT", price, buy_qty))
                remaining -= buy_qty

        if state.timestamp == 99900 and curr_pepper_position > 0:
            orders.append(Order("INTARIAN_PEPPER_ROOT", best_bid, -curr_pepper_position))

class OsmiumTrader:
    def take_obvious_trades(best_bid, bid_volume, best_ask, ask_volume, mid_price, orders, curr_osmium_position):
        if best_bid > mid_price and curr_osmium_position >= bid_volume:
            orders.append(Order("ASH_COATED_OSMIUM", best_bid, -bid_volume))

        if best_ask < mid_price and curr_osmium_position <= MAX_VOLUME - ask_volume:
            orders.append(Order("ASH_COATED_OSMIUM", best_ask, ask_volume))

    def make_better_trades(best_bid, best_ask, mid_price, size, orders, curr_osmium_position):
        if (best_bid + 1 < mid_price) and curr_osmium_position <= MAX_VOLUME - size:
            orders.append(Order("ASH_COATED_OSMIUM", best_bid + 1, size))

        if (best_ask - 1 > mid_price) and curr_osmium_position >= size:
            orders.append(Order("ASH_COATED_OSMIUM", best_ask - 1, -size))

class Trader:
    def get_best_bid(self, order_depth):
        if not order_depth.buy_orders:
            return None, 0
        best_bid = max(order_depth.buy_orders.keys())
        return best_bid, order_depth.buy_orders[best_bid]

    def get_best_ask(self, order_depth):
        if not order_depth.sell_orders:
            return None, 0
        best_ask = min(order_depth.sell_orders.keys())
        return best_ask, order_depth.sell_orders[best_ask]

    def run(self, state: TradingState) -> tuple[dict[Symbol, list[Order]], int, str]:
        result = {}
        conversions = 0
        trader_data = ""

        pepper_orders = []
        osmium_orders = []

        curr_osmium_position = state.position.get("ASH_COATED_OSMIUM", 0)
        curr_pepper_position = state.position.get("INTARIAN_PEPPER_ROOT", 0)

        # ================= OSMIUM =================
        order_depth = state.order_depths.get("ASH_COATED_OSMIUM")
        if order_depth:
            best_bid, bid_volume = self.get_best_bid(order_depth)
            best_ask, ask_volume = self.get_best_ask(order_depth)

            if best_bid is not None and best_ask is not None:
                mid_price = (best_bid + best_ask) / 2
                spread = best_ask - best_bid

                inventory_factor = 1 - abs(curr_osmium_position) / MAX_VOLUME
                osmium_trade_size = int(spread * 5 * inventory_factor)

                osmium_trade_size = max(
                    1,
                    min(osmium_trade_size, bid_volume, -ask_volume, 20)
                )

                OsmiumTrader.take_obvious_trades(
                    best_bid, bid_volume,
                    best_ask, ask_volume,
                    mid_price,
                    osmium_orders,
                    curr_osmium_position
                )

                OsmiumTrader.make_better_trades(
                    best_bid, best_ask,
                    mid_price,
                    osmium_trade_size,
                    osmium_orders,
                    curr_osmium_position
                )

                result["ASH_COATED_OSMIUM"] = osmium_orders

        # ================= PEPPER =================
        order_depth = state.order_depths.get("INTARIAN_PEPPER_ROOT")
        if order_depth:
            best_bid, bid_volume = self.get_best_bid(order_depth)

            if best_bid is not None:
                PepperTrader.long_only(
                    order_depth,
                    state,
                    best_bid,
                    pepper_orders,
                    curr_pepper_position
                )

                result["INTARIAN_PEPPER_ROOT"] = pepper_orders

        logger.flush(state, result, conversions, trader_data)
        return result, conversions, trader_data
```
